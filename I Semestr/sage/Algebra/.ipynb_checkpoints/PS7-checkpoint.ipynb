### Zadanie 1
---
Rozwiąż równanie w liczbach zespolonych:

a) $|z|+z=8+4i;$

b) $z^2+\overline{3z}=0.$

```python
#a)
x,y = var('x,y')
assume(x,'real',y,'real')
z=x+i*y
m(x,y)=sqrt(x^2+y^2)
eq = m(x,y)+z==8+4*i
show(eq)
ukl = [eq.lhs().real_part()==eq.rhs().real_part(),eq.lhs().imag_part()==eq.rhs().imag_part()]
show(ukl)
solve(ukl,[x,y])
```

```python
#b)
x,y = var('x,y')
assume(x,'real',y,'real')
z=x+i*y

eq = z^2+(3*z).conjugate()==0
show(eq)
ukl = [eq.lhs().real_part()==eq.rhs().real_part(),eq.lhs().imag_part()==eq.rhs().imag_part()]
show(ukl)
show(solve(ukl,[x,y]))
```

### Zadanie 2
---
Rozwiąż równanie w liczbach zespolonych:

a) $z^2+z+1=0;$

b) $z^4+(15+7i)z^2+(8-15i)=0.$

Zaznacz rozwiązania na płaszczyźnie.

```python
#a)
z = var('z')
rozw = solve(z^2+z+1==0,z)
show(rozw)
points([rozw[j].rhs() for j in range(len(rozw))],size=40)
```

```python
list_plot([rozw[j].rhs() for j in range(len(rozw))],size=40)
```

```python
#b)
z = var('z')
rozw = solve(z^4+(15+7*i)*z^2+(8-15*i)==0,z)
show(rozw)
points([rozw[j].rhs() for j in range(len(rozw))],size=40)
```

### Zadanie 3
---
Narysuj na płaszczyźnie:

siedmiokąt foremny, którego jeden z wierzchołków leży w punkcie $(1,1)$;

```python
z = var('z')
rozw = solve(z^7==(1+i)^7,z)
show(rozw)
p1 = points([rozw[j].rhs() for j in range(len(rozw))],size=40,color='red')
p2 = polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false)
p3 = circle((0,0),abs(rozw[0].rhs()),color='green')
p1+p2+p3
```

### Zadanie 4
---
Na płaszczyźnie zespolonej narysuj zbiory:

a) $A = \{z\in\mathbb{C}:\ |z|=1\};$ 


```python
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
region_plot(m(x,y)==1,(x,-2,2),(y,-2,2))
```

b) $B = \{z\in\mathbb{C}:\ |z|\leqslant 2\};$


```python
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
region_plot(m(x,y)<=2,(x,-2,2),(y,-2,2),incol='lightblue',bordercol='grey',borderstyle='solid')
```

c) $C = \{z\in\mathbb{C}:\ |z-1-i|<1\};$ 


```python
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
region_plot(m(x-1,y-1)<1,(x,-2,2),(y,-2,2),incol='lightblue',bordercol='grey',borderstyle='dashed')
```

d) $D = \{z\in\mathbb{C}:\ 2<|z|<3\};$

```python
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
region_plot([m(x,y)<3,m(x,y)>2],(x,-4,4),(y,-4,4),incol='lightblue',bordercol='grey',borderstyle='dashed')
```

e) $E = \{z\in\mathbb{C}:\ 1\leqslant |z-2i|<2\};$  

```python
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
p1 = region_plot([m(x,y-2)<2,m(x,y-2)>=1],(x,-4,4),(y,-4,4),incol='lightblue')
p2 = region_plot(m(x,y-2)==2,(x,-4,4),(y,-4,4),bordercol='grey',borderstyle='dashed')
p3 = region_plot(m(x,y-2)==1,(x,-4,4),(y,-4,4),bordercol='grey',borderstyle='solid')
p1+p2+p3
```
