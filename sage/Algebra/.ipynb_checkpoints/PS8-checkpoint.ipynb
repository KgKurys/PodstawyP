### Zadanie 1
---
Narysuj na płaszczyźnie:

kwadrat, którego jeden z wierzchołków leży w punkcie $(3,5)$.

Zakładamy, że środek wielokąta znajduje się w punkcie $(2,2)$.

```python
srodek = 2+2*i
wierzcholek = 3+5*i
```

```python
nowy_wierzcholek = wierzcholek-srodek
show(nowy_wierzcholek)
```

```python
z = var('z')
rozw = solve(z^4==(nowy_wierzcholek)^4,z)
show(rozw)
```

```python
polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false)
```

```python
wierzcholki = [rozw[j].rhs()+srodek for j in range(len(rozw))]
show(wierzcholki)
```

```python
polygon(wierzcholki,fill=false)
```

```python
p1 = polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false,color='grey')
p2 = polygon(wierzcholki,fill=false,color='red')
p3 = points([srodek, wierzcholek],color='black',size=30)
p1+p2+p3
```

### Zadanie 2

Wyznaczyć obraz kwadratu z zadania 1 przy przekształceniu płaszczyzny danym wzorem $$\varphi\left(\left(\begin{array}{rr}x\\ y\end{array}\right)\right)=A\cdot\left(\begin{array}{rr}x\\ y\end{array}\right),$$
gdzie $A=\left[\begin{array}{rr}2&-0.3\\ 0.7&0.3\end{array}\right]$.

```python
A = matrix(2,2,[2,-0.3,0.7,0.3])
show(A)
```

```python
lista_wierzcholkow = [matrix(2,1,[wierzcholki[j].real_part(),wierzcholki[j].imag_part()]) for j in range(len(wierzcholki))]
show(lista_wierzcholkow)
```

```python
po_przeksztalceniu = [A*lista_wierzcholkow[j] for j in range(len(wierzcholki))]
show(po_przeksztalceniu)
```

```python
polygon([(po_przeksztalceniu[j][0,0],po_przeksztalceniu[j][1,0]) for j in range(len(rozw))],fill=false)
```

```python
k1 = polygon([(po_przeksztalceniu[j][0,0],po_przeksztalceniu[j][1,0]) for j in range(len(rozw))],fill=false, color='red')
k2 = polygon(wierzcholki,fill=false)
k1+k2
```

### Zadanie 3
---
Trójkąt foremny o jednym z wierzchołków w punkcie $(-2,5)$ i środku w początku układu współrzędnych obrócić o kąt $\frac{\pi}{10}$ zgodnie z ruchem wskazówek zegara.

```python
z = var('z')
wierzcholki = solve(z^3==(-2+5*i)^3,z)
show(wierzcholki)
```

```python
p1 = polygon([wierzcholki[j].rhs() for j in range(len(wierzcholki))],fill=false,color='blue')
p2 = points([(0,0),(-2,5)],color='black',size=30)
p1+p2
```

```python
obrot = [wierzcholki[j].rhs()*exp(-(pi/10)*i) for j in range(len(wierzcholki))]
polygon(obrot,color='red',fill=false)
```

```python
p1 = polygon([wierzcholki[j].rhs() for j in range(len(wierzcholki))],fill=false,color='blue')
p2 = points([(0,0),(-2,5)],color='black',size=30)
p3 = polygon(obrot,color='red',fill=false)
p1+p2+p3
```

### Zadanie 4
---
Pięciokąt o jednym z wierzchołków w punkcie $(0,5)$ i środku w w punkcie $(-1,2)$ obrócić względem środka figury o kąt $\frac{\pi}{6}$ przeciwnie do ruchu wskazówek zegara.

```python
srodek = -1+2*i
wierzcholek = 0+5*i
```

```python
nowy_wierzcholek = wierzcholek-srodek
show(nowy_wierzcholek)
```

```python
z = var('z')
rozw = solve(z^5==(nowy_wierzcholek)^5,z)
show(rozw)
polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false,color='blue')
```

```python
obrot = [rozw[j].rhs()*exp(pi/6*i) for j in range(len(rozw))]
show(obrot)
polygon(obrot,color='red',fill=false)
```

```python
wierzcholki = [rozw[j].rhs()+srodek for j in range(len(rozw))]
show(wierzcholki)
```

```python
polygon(wierzcholki,fill=false)
```

```python
odp = [obrot[j]()+srodek for j in range(len(rozw))]
show(odp)
polygon(odp,fill=false)
```

```python
p1 = polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false,color='grey')
p2 = polygon(obrot,color='lightblue',fill=false)
p3 = points([(0,5),(-1,2)],color='black',size=30)
p4 = polygon(wierzcholki,fill=false, color='blue')
p5 = polygon(odp,fill=false,color='red')
p1+p2+p3+p4+p5
```

```python
p3 = points([(0,5),(-1,2)],color='black',size=30)
p4 = polygon(wierzcholki,fill=false, color='blue')
p5 = polygon(odp,fill=false,color='red')
p3+p4+p5
```

$$A=\{z\in\mathbb{C}:\ |z|<10\quad\mbox{i}\quad |z-2i|>3\quad\mbox{i}\quad\mathrm{Re\,}z<1\}.$$

2. Na płaszczyźnie zespolonej narysuj zbiór:

$$B=\{z\in\mathbb{C}:\ |z|<5\quad\mbox{i}\quad |z-2|>3\quad\mbox{i}\quad\mathrm{Im\,}z>-1\}.$$

3. Narysuj na płaszczyźnie siedmiokąt foremny, którego jeden z wierzchołków leży w punkcie $(-2,5)$. Zakładamy, że środek wielokąta znajduje się w początku układu współrzędnych.


4. Narysuj na płaszczyźnie dziesięciokąt foremny, którego jeden z wierzchołków leży w punkcie $(4,-3)$. Zakładamy, że środek wielokąta znajduje się w początku układu współrzędnych.

```python
1#
forget()
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
p1 = region_plot([m(x,y)<10 , m(x,y-2)>3],x<1,(x,-11,11),(y,-11,11),incol='lightblue')


p2 = region_plot(m(x,y)==10,(x,-11,11),(y,-11,11),bordercol='blue',borderstyle='solid')
p3 = region_plot(m(x,y-2)==3,(x,-11,11),(y,-11,11),bordercol='red',borderstyle='solid')
p4 = region_plot(x<1,(x,-11,11),(y,-11,11),bordercol='black',borderstyle='solid')
p1+p2+p3+p4
```

```python
#3
srodek = 0+0*i
wierzcholek = -2+5*i
nowy_wierzcholek = wierzcholek-srodek
show(nowy_wierzcholek)
z = var('z')
rozw = solve(z^7==(nowy_wierzcholek)^7,z)
show(rozw)
polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false)
```

```python
#4
srodek = 0+0*i
wierzcholek = 4+(-3)*i
nowy_wierzcholek = wierzcholek-srodek
show(nowy_wierzcholek)
z = var('z')
rozw = solve(z^7==(nowy_wierzcholek)^7,z)
show(rozw)
polygon([rozw[j].rhs() for j in range(len(rozw))],fill=false)
```
