```python
1
x,y = var('x,y')
assume(x,'real',y,'real')
m(x,y) = sqrt(x^2+y^2)
p1 = region_plot([m(x,y)<=3],(x,-10,10),(y,-10,10),incol='blue')
p2 = region_plot([m(x+2,y-3)>=5],(x,-10,10),(y,-10,10),incol='grey')
p3 = region_plot([x>=1], (x,-10,10),(y,-10,10),incol='green')
p1+p2+p3
odp= region_plot([m(x,y)<=3,m(x+2,y-3)>=5,x>=1],(x,-5,5),(y,-5,5), incol='blue', bordercol='black', borderstyle='solid')
odp
```
