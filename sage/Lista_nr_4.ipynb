```python
forget()
f(x) = x^2*sqrt(36-x^2)
p1 = plot(36-x^2,(x,-10,10),ymin=-50,ymax=50)
show(p1)
#df = <-6,6>

fp(x) = diff(f(x),x)
show("fp(x)=",fp(x).full_simplify().factor())

#Dfp = (-6,6)
fp_licz(x) = -3*(x^2-24)*x
rozw1 = solve(fp_licz(x)>0,x)
show(rozw1)

#f'(x) jest wieksze od zera dla x z przedzialow:(-6, -2*sqrt(6))u(0,2*sqrt(6)) - funckja f jest rosnaca
#f'(x)<0 dla x z przedzialow: (-2*sqrt(6);0)u(2*sqrt(6);6) - funckja f jest malejaca
```
